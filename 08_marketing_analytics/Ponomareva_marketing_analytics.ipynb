# Маркетинговая аналитика в Яндекс.Афише

Целью настоящего исследования является подсчёт ключевых бизнес-показателей и анализ пользовательского поведения на сайте Яндекс.Афиши. Материалом для исследования служат данные о посещениях, заказах и рекламных расходах за период с июня 2017 года по конец мая 2018 года. Результаты будут использованы для оптимизации расходов на маркетинговые кампании.

Задачи исследования:
1. Изучить данные об активности использования сервиса, посчитать:
    - число уникальных пользователей за день, неделю, месяц, 
    - среднее количество посещений сайта в день, 
    - среднюю продолжительность сессии, 
    - коэффициент удержания;
2. Выяснить, когда клиенты делают первые покупки на сайте и сколько в среднем совершают заказов;
3. Рассчитать величину среднего чека и LTV каждого клиента;
4. Изучить, сколько денег тратится на рекламные кампании, и выявить самые выгодные источники трафика.

### Продуктовые метрики

#### Число активных пользователей (DAU, WAU, MAU)

Рассчитаем метрики DAU, WAU и MAU: количество уникальных пользователей в день, неделю и месяц.

```python
#извлечём из даты начала сессии день, номер недели, месяц и год
visits['day'] = visits['start_ts'].dt.date
visits['week'] = visits['start_ts'].dt.week
visits['month'] = visits['start_ts'].dt.month
visits['year'] = visits['start_ts'].dt.year
```

```python
#вычислим количество активных пользователей 
dau_total = visits.groupby('day').agg({'uid': 'nunique'})
wau_total = visits.groupby(['year', 'week']).agg({'uid': 'nunique'})
mau_total = visits.groupby(['year', 'month']).agg({'uid': 'nunique'})
```

```python
#вычислим среднее значение метрики за весь период и построим график её временной динамики
def active_users(metrics, time):
    display('Среднее число уникальных пользователей в ' + time +': {}'.format(round(metrics.mean()[0])))
    metrics.plot(grid=True, legend=False, figsize=(16, 4))
    plt.title('Изменение числа активных пользователей в ' + time, fontsize=16)
    plt.xlabel('Время')
    plt.ylabel('Количество пользователей')
    plt.show()
```

```python
active_users(dau_total, 'день')
```

Мы видим, что в среднем график колеблется между 500 и 1500 уникальных посетителей в день, но есть два дня (в ноябре 2017 года и марте 2018 года), которые сильно выбиваются из общей динамики. Посмотрим, что происходило в эти дни.

```python
dau_total.query('uid == uid.max() | uid == uid.min()')
#узнаем даты с наибольшим и наименьшим числом посетителей
```

```python
visits.query('"2017-11-25 00:00:00" > start_ts > "2017-11-23 23:59:59"').sample(10)
```

24 ноября 2017 года на сайте было много пользователей, пришедших из разных источников. Возможно, незадолго до 24 ноября было запущено сразу несколько успешных рекламных кампаний или на Яндекс.Афише в этот день была большая пятничная распродажа.

```python
visits.query('"2018-04-01 00:00:00" > start_ts > "2018-03-30 23:59:59"')
```

31 марта 2018 года была зафиксирована только одна сессия продолжительностью в 1 минуту. Вероятно, в этот день счётчик посещений сломался, поэтому данные неполные.

```python
active_users(wau_total, 'неделю')
```

Количество активных пользователей сайта начало уверенно расти с 36 недели 2017 года, то есть в начале сентября. Максимальное число посетителей - 10586 человек - было зафиксировано на неделе с 20 по 26 ноября 2017 года. С декабря 2017 по конец марта 2018 года количество посетителей составляло от 6 до 8 тысяч человек в неделю, с 13 недели 2018 года (26 марта - 1 апреля, в эту неделю входит вышеизученный день с единственным посетителем) наблюдается снижение числа уникальных пользователей. Для оценки степени лояльности аудитории дополнительно рассчитаем метрику вовлечённости - Sticky Factor.

```python
display('Sticky Factor за неделю: {:.2%}'.format(dau_total.mean()[0] / wau_total.mean()[0]))
```

```python
active_users(mau_total, 'месяц')
```

```python
display('Sticky Factor за месяц: {:.2%}'.format(dau_total.mean()[0] / mau_total.mean()[0]))
```

Данный график в целом похож на предыдущий: наблюдаем рост числа активных пользователей с сентября 2017 года, пик посещений в ноябре, стабильное снижение метрики с декабря 2017 года по март 2018 года и резкое падение активности в апреле-мае 2018 года. Регулярность использования сайта достаточно низкая: Sticky Factor за месяц составляет около 4%.

#### Среднее количество посещений в день

Теперь изучим общее число сессий. Посчитаем, сколько раз за день пользователи в среднем заходят на сайт, и посмотрим на изменение этой метрики во времени.

```python
total_visits = visits.groupby('day')['start_ts'].count()
display('Среднее число посещений в день: {}'.format(round(total_visits.mean())))
total_visits.plot(grid=True, legend=False, figsize=(16, 4))
plt.title('Среднее число посещений в день', fontsize=16)
plt.xlabel('Дата')
plt.ylabel('Количество посещений');
```

Среднее число всех посещений за день составляет 987, что чуть выше среднего DAU (908 пользователей). Максимальное число сессий за день (4042) зафиксировано 24 ноября 2017 года, и оно также выше количества уникальных пользователей (3319), то есть некоторые посетители возвращались на сайт неоднократно. #### Retention Rate

Чтобы понять, как долго посетители продолжают пользоваться сервисом после первого обращения, посчитаем Retention Rate, он же коэффициент удержания. Для этого применим когортный анализ. Поскольку у нас в распоряжении данные за год, будет логично выделить когорты по месяцам.

```python
#выделим самую раннюю дату, в которую была замечена активность пользователя
first_visit = visits.groupby('uid')['start_ts'].min()
first_visit.name = 'first_visit'
```

```python
visits = visits.join(first_visit, on='uid') #объединим с общим датафреймом
```

```python
visits_grouped = visits[['uid', 'start_ts', 'first_visit']].reset_index() #выделим нужные столбцы
```

```python
visits_grouped['first_visit_month'] = visits_grouped['first_visit'].astype('datetime64[M]')
visits_grouped['month'] = visits_grouped['start_ts'].astype('datetime64[M]')
```

```python
visits_grouped['cohort_lifetime'] = (visits_grouped['month'] - visits_grouped['first_visit_month'])
#посчитаем "срок жизни" когорты
```

```python
#переведем "срок жизни" из дней в месяцы
visits_grouped['cohort_lifetime'] = visits_grouped['cohort_lifetime'] / np.timedelta64(1, 'M')
visits_grouped['cohort_lifetime'] = visits_grouped['cohort_lifetime'].round().astype('int')
```

```python
cohorts = visits_grouped.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}) \
.reset_index() #посчитаем количество уникальных пользователей в каждой когорте
```

Чтобы вычислить Retention Rate, нужно получить число пользователей, изначально бывших в когорте, а потом разделить на него число пользователей в каждый следующий период (в нашем случае, месяц).

```python
cohort_users = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month', 'uid']]
cohort_users = cohort_users.rename(columns={'uid':'cohort_users'})
#выделим пользователей, которые были в когорте изначально
```

```python
cohorts = cohorts.merge(cohort_users, on='first_visit_month') #объединим таблицы
```

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users'] #посчитаем retention
```

```python
cohorts['first_visit_month'] = cohorts['first_visit_month'].dt.strftime('%Y-%m') 
#оставим в обозначении когорты только год и месяц, чтобы красиво визуализировать данные
```

```python
#создадим сводную таблицу и построим тепловую карту
retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',    
    values='retention',    
    aggfunc='sum', 
) 

sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Retention Rate пользователей Яндекс.Афиши')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.ylabel('Когорта')
plt.xlabel('Срок жизни когорты');
```

Как полагается, Retention Rate на нулевой месяц существования когорты составляет 100%, а вот потом показатели становятся низкими: лучший результат на первый месяц существования показала когорта сентября 2017 года с коэффициентом удержания 8,5%. Самая позитивная динамика в целом у когорты июня 2017 года: в этой когорте Retention показал небольшой рост на четвёртый, пятый и одиннадцатый месяц lifetime, показатель за последний месяц (4,5%) сравним с показателем когорты апреля 2018 года (4,8%), для которой это первый lifetime, тогда как у остальных когорт Retention в мае 2018 года составил от 1,9% до 2,7%.

```python
#посчитаем средний Retention Rate на второй месяц "жизни" когорт
avg_retention = cohorts.query('cohort_lifetime == 1')['retention'].mean()
display('Средний Retention Rate на второй месяц "жизни" когорт составляет {:.1%}'.format(avg_retention))
```

Несмотря на наличие постоянных клиентов, в целом малая часть пользователей возвращается на сайт повторно. Возможно, у большинства посетителей не так часто возникает потребность в покупке билетов в кино или на концерт, либо они пользуются для этого другими сервисами.

#### Среднее количество покупок на одного покупателя 

Рассчитаем количество покупок на одного покупателя. Для вычисления среднего показателя применим когортный анализ, выделим когорты по первому месяцу покупки.

```python
#выделим месяц из даты заказов
orders['first_order_month'] = orders['first_order'].astype('datetime64[M]')
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
```

```python
#посчитаем количество новых покупателей за каждый месяц
cohort_sizes = (orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
#посчитаем количество заказов по когортам за каждый месяц
orders_n = orders.groupby(['first_order_month', 'order_month']).agg({'buy_ts':'count'}).reset_index()
orders_n.columns = ['first_order_month', 'order_month', 'n_orders']
```

#### Средний LTV

Рассчитаем средний показатель lifetime value — валовой прибыли с каждого клиента. Снова применим когортный анализ.

```python
#добавим рассчитанный ранее размер когорт в исходный датафрейм
orders = orders.merge(cohort_sizes, on='first_order_month', how='left')
```

```python
#посчитаем срок жизни когорты
orders['cohort_lifetime'] = (orders['order_month'] - orders['first_order_month'])
orders['cohort_lifetime'] = orders['cohort_lifetime'] / np.timedelta64(1, 'M')
orders['cohort_lifetime'] = (orders['cohort_lifetime'].round().astype('int'))
```

За расчётный период примем 6 месяцев. Исключим когорты, которые "прожили" меньше полугода.

```python
orders_for_ltv = orders.query('first_order_month < "2018-01-01"').reset_index()
```

```python
total_revenue = total_revenue.rename(columns={'revenue':'total_revenue'})
```

```python
#добавим значения общей выручки в датафрейм
orders_for_ltv = orders_for_ltv.merge(total_revenue, on=['first_order_month', 'order_month'], how='left')
```

```python
orders_for_ltv['ltv'] = orders_for_ltv['total_revenue'] / orders_for_ltv['n_buyers']
```

```python
#оставим в названии когорт только год и месяц
orders_for_ltv['first_order_month'] = orders_for_ltv['first_order_month'].dt.strftime('%Y-%m')
```

```python
#составим сводную таблицу и построим тепловую карту
ltv = orders_for_ltv.pivot_table(index='first_order_month', columns='cohort_lifetime', values='ltv')

sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Средний LTV пользователей Яндекс.Афиши')
# поскольку LTV — накопительная метрика, к каждому следующему месяцу прибавим результат предыдущего
sns.heatmap(ltv.cumsum(axis=1), annot=True, fmt='.3', linewidths=1, linecolor='gray')
plt.ylabel('Когорта')
plt.xlabel('Срок жизни когорты');
```

График показывает, что больше всего выручки сайту принесли пользователи из когорты сентября 2017 года: на четвёртый месяц её существования, в декабре 2017 года, LTV существенно (почти на 4 у.е.) вырос - похоже, именно в этой когорте наиболее активно пользовались предновогодними предложениями. В остальных когортах LTV растёт незначительно - в среднем на 0,5 у.е. каждый месяц. Самые низкие показатели в октябрьской и ноябрьской когортах - прибавляется меньше 0,2 у.е. 

```python
display('Средний LTV за 6 месяцев после первой покупки: {:.2f} у.е.'.format(ltv.cumsum(axis=1).mean(axis=0)[5]))
```

#### Средний CAC

Вычислим customer acquisition cost — стоимость привлечения одного покупателя. Исследуем эту метрику для всех затрат в целом и по каждому источнику трафика. Поскольку один и тот же пользователь мог заходить на сайт из нескольких источников, выделим для каждого первый источник перехода и будем рассматривать его в качестве главного.

```python
main_source_id = visits.groupby(['uid', 'first_visit'])['source_id'].first().reset_index()
```

```python
#добавим информацию про главный источник в таблицу с заказами
orders = orders.merge(main_source_id, on=['uid', 'first_visit'])
```

#### ROMI

Чтобы понять, какие источники трафика приносят сайту наибольший доход, рассчитаем показатель ROMI - окупаемость маркетинговых инвестиций. Чтобы вычислить средний ROMI для одного покупателя, необходимо разделить валовую прибыль на затраты, то есть LTV на CAC. Пересчитаем метрики, выделив когорты не только по месяцам, но и по рекламным источникам.

```python
#посчитаем количество покупателей в когорте
cohorts_size = (orders.groupby(['first_order_month','source_id']).agg({'uid': 'nunique'}).reset_index())
```

```python
cohorts_size = cohorts_size.rename(columns={'uid':'cohort_size'})
```

```python
#посчитаем выручку по источникам и месяцам
revenue_per_cohorts = orders.groupby(['source_id', 'first_order_month', 'order_month'], as_index=False)\
                        .agg({'revenue':'sum'})
```

```python
#найдём срок жизни когорты для каждого заказа
revenue_per_cohorts['age'] = (revenue_per_cohorts['order_month'] - revenue_per_cohorts['first_order_month'])
revenue_per_cohorts['age'] = revenue_per_cohorts['age'] / np.timedelta64(1, 'M')
revenue_per_cohorts['age'] = (revenue_per_cohorts['age'].round().astype('int'))
```

```python
#добавим информацию про размер когорт
revenue_per_cohorts = revenue_per_cohorts.merge(cohorts_size, on=['source_id', 'first_order_month'])
```

```python
#вычислим ltv
revenue_per_cohorts['ltv'] = revenue_per_cohorts['revenue'] / revenue_per_cohorts['cohort_size']
```

```python
#добавим информацию про маркетинговые затраты по источникам
revenue_per_cohorts = revenue_per_cohorts.merge(costs_per_source, 
                        left_on=['source_id', 'first_order_month'], right_on=['source_id', 'month'])
```

```python
#вычислим cac
revenue_per_cohorts['cac'] = revenue_per_cohorts['costs'] / revenue_per_cohorts['cohort_size']
```

```python
#вычислим romi
revenue_per_cohorts['romi'] = revenue_per_cohorts['ltv'] / revenue_per_cohorts['cac']
```

```python
#создадим сводную таблицу
output = revenue_per_cohorts.pivot_table(index='source_id', columns='age', values='romi')
```

```python
#построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Средний ROMI за одинаковые периоды жизни когорт с учётом источников трафика')
sns.heatmap(output.cumsum(axis=1), annot=True, fmt='.3', linewidths=1, linecolor='gray')
plt.ylabel('Источник')
plt.xlabel('Срок жизни когорты');
```

Самыми прибыльными рекламными источниками являются №1 и №9: они окупились уже в первый месяц. Неплохой результат у источника №2: он окупился на четвёртый месяц. Источник №5 окупился на двенадцатый месяц. Остальные источники убыточные, самыми невыгодными являются вложения в источник №3, который одновременно и самый дорогой.

```python
#вычислим средний ROMI всех когорт за каждый месяц их жизни
romi = pd.DataFrame(output.cumsum(axis=1).mean(axis=0))
romi = romi.rename(columns={0:'ROMI'})
display(romi.style.background_gradient(cmap='Greens'))
```

В среднем инвестиции по всем источникам окупаются после четырёх месяцев существования когорты (ROMI 1,05), наибольшая прибыль достигается на 10-11 месяц (ROMI 1,5).
